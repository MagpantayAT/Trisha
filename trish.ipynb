### import needed libraries

```python
import pandas as pd
from pandas import Series, DataFrame
```

### read excel

```python
#
# read natin un excel then store sa variable named 'df'
# df means dataframe sa pandas naming convention lang
#
df = pd.read_excel('data.xlsx')

# dispaly lang natin
df
```

### Split cell into rows

```python
#
# Col2 is the column to be splitted
# Col1 is your unique values
#
df_split = pd.concat([Series(row['Col1'], row['Col2'].split(',')) for _, row in df.iterrows()]).reset_index()

# tignan lang nain kung nag split
df_split
```

```python
#
# Rename lang natin un columns back to original
# result column is named 'index' which is the Col2 originally
# and 0 for Col1
#

df_split = df_split.rename(columns={'index': 'Col2', 0:'Col1'})

# dispaly lang natin
df_split
```

```python
#
# after split un cell into rows, balik na natin sa original dataframe
# 'Col1', 'Col3' is un mga remaining columns na need natin
# 'Col1' is un common column natin sa pag merge
#
res = df_split.merge(df[['Col1', 'Col3']], on="Col1", how = 'inner')

# display lang
res
```

```python
#
# ulitin lang for suceeding columns
#
df_split = pd.concat([Series(row['Col1'], row['Col3'].split(',')) for _, row in df.iterrows()]).reset_index()

df_split = df_split.rename(columns={'index': 'Col3', 0:'Col1'})

# dispaly lang natin
df_split
```

```python
# merge natin sa previous dataframe 'res'
res2 = df_split.merge(res[['Col1', 'Col2']], on="Col1", how = 'inner')

res2
```

```python
# rearrange mo lang un columns as needed kung gusto mo
final = res2[['Col1', 'Col2', 'Col3']]

final
```

```python
#
# save na natin
#
final.to_excel('trish.xlsx')
```
